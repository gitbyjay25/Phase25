# apply + lamda
lambda arguments: expression


*   Categorize Traffic Density
*  Check if Speed Is Low (below limit)




```python
df["traffic_density"]=df["Vehicles"].apply(lambda x: "high" if x>27 else "medium" if x >=23 else "low")
```

```python
df
```

```python
df["Speed_type"]=df["Avg_Speed"].apply(lambda x: "high" if x>42.5 else "medium" if x >=32 else "low")
```

```python
df
```
